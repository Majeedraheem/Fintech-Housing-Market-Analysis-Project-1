### 1a. What are the trends for Ontario real estate for 2005-2023? (Yearly)

```python
# Import the required libraries and dependencies
import pandas as pd
from pathlib import Path
from bokeh.models.formatters import NumeralTickFormatter
from bokeh.models.formatters import DatetimeTickFormatter
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
#Import Federal and Provincial Data
ontario_df = pd.read_csv(Path("Resources/ONTARIO.csv"))
```

```python
#Set date as index 
ontario_df = ontario_df.set_index("Date")
```

```python
#Show graph of Ontario Composite HPI for 2005-2023
ontario_df["Composite_HPI_SA"].dropna().hvplot.line(
    label="Ontario Composite House Price Index 2005-2023",
    xlabel="Date",
    ylabel="Composite House Price Index",
    rot=90,
    width=1400,
    height = 500
).opts(
    yformatter=NumeralTickFormatter(format="0,0"),
    fontsize={
        'title': 20, 
        'labels': 14, 
        'xticks': 5, 
        'yticks': 10,
    }
)
```

![Composite House Price Index (with text)](Images/CompositeHousePriceIndex.png "Composite House Price Index")

```python
#Show graph of all House Price Indicies
ontario_df.dropna().loc[:, ['Composite_HPI_SA', 'Single_Family_HPI_SA', 'One_Storey_HPI_SA', 'Two_Storey_HPI_SA', 'Townhouse_HPI_SA', 'Apartment_HPI_SA']].hvplot.line(
    label="Ontario HPI 2005-2023",
    xlabel="Date",
    ylabel="Composite Benchmark SA",
    rot=90,
    width=1400,
    height = 500
).opts(
    yformatter=NumeralTickFormatter(format="0,0"),
    fontsize={
        'title': 20, 
        'labels': 14, 
        'xticks': 5, 
        'yticks': 10,
    }
)
```

![Ontario HPI (with text)](Images/OntarioHPI_with_text.png "Ontario HPI")
